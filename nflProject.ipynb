```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/BrianBehnke/DS-Sprint-01-Dealing-With-Data/master/NFLOutput%20Extract.csv'
nfl_data=pd.read_csv(url)
nfl_data.head()
```

```python
nfl_data.shape
```

```python
nfl_data.describe
```

```python
nfl_data.sample(10)
```

```python
nfl_data_no_nan = nfl_data.fillna(0)
nfl_data_no_nan.sample(10)
```

```python
nfl_data_no_nan.head()
```

```python
nfl_data_no_nan['Position'].value_counts()
```

```python
nfl_data_no_nan['Team'].value_counts()
```

```python
nfl_data_no_nan.dtypes
```

```python
nfl_data_no_nan['Position'] = nfl_data_no_nan['Position'].astype(str)
nfl_data_no_nan.dtypes
```

```python
fig, ax1 = plt.subplots(figsize=(20, 12))
plt.title('Scatter plot of Salary vs Position for NFL Data')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.scatter(nfl_data_no_nan['Position'], nfl_data_no_nan['Salary'])
plt.yscale('log')
plt.show()
```

```python
# That looks much nicer than the pie chart
# Lets try a box plot
fig, ax1 = plt.subplots(figsize=(6, 6))
plt.title('Boxplot of Slary for NFL Data')
plt.ylabel('Salary')

plt.boxplot(nfl_data_no_nan['Salary'])
plt.show()
```

```python
ax = nfl_data_no_nan[['Team','Salary']].plot(kind='bar', title ="Team per salary", figsize=(30, 10), legend=True, fontsize=12)
ax.set_xlabel("Team", fontsize=12)
ax.set_ylabel("Annual salary", fontsize=12)
plt.yscale('log')
plt.show()
```

```python
nfl_data_no_nan.describe()
```

```python
#nfl_scorers = []
nfl_scorers = nfl_data_no_nan.loc[nfl_data_no_nan['Pts'] >= 10]
```

```python
nfl_scorers.head().T
```

```python
nfl_scorers.describe()
```

```python
ax = nfl_scorers[['Pts','Salary']].plot(kind='bar', title ="Player per salary", figsize=(30, 10), legend=True, fontsize=12)
ax.set_ylabel("Player", fontsize=12)
ax.set_xlabel("Annual salary", fontsize=12)
#plt.yscale('log')
plt.show()
```

```python
#nfl_scorers = nfl_scorers.set_index(['Player'])
nfl_scorers.head().T
```

```python
nfl_scorers.head()
```

```python
nfl_scorers = nfl_scorers.drop('2PM', axis=1)
nfl_scorers = nfl_scorers.drop('Age', axis=1)
nfl_scorers = nfl_scorers.drop('All TD', axis=1)
nfl_scorers = nfl_scorers.drop('FGA', axis=1)
nfl_scorers = nfl_scorers.drop('FGM', axis=1)
nfl_scorers = nfl_scorers.drop('Fbl TD', axis=1)
nfl_scorers = nfl_scorers.drop('GS', axis=1)
nfl_scorers = nfl_scorers.drop('Games', axis=1)
nfl_scorers = nfl_scorers.drop('Int TD', axis=1)
nfl_scorers = nfl_scorers.drop('Kr Td', axis=1)
nfl_scorers = nfl_scorers.drop('Number of Records', axis=1)
nfl_scorers = nfl_scorers.drop('Oth TD', axis=1)
nfl_scorers = nfl_scorers.drop('Pr Td', axis=1)
nfl_scorers = nfl_scorers.drop('Pts/G', axis=1)
nfl_scorers = nfl_scorers.drop('Rec TD', axis=1)
nfl_scorers = nfl_scorers.drop('Rsh TD', axis=1)
nfl_scorers = nfl_scorers.drop('Sfty', axis=1)
nfl_scorers = nfl_scorers.drop('XPA', axis=1)
nfl_scorers = nfl_scorers.drop('XPM', axis=1)
```

```python
nfl_scorers.head()
```

```python
nfl_scorers
```

```python
nfl_scorers.describe()
```

```python
ax = nfl_scorers[['Player','Salary']].plot(kind='bar', title ="Player per salary", figsize=(30, 10), fontsize=20)
ax.set_ylabel("Player", fontsize=20)
ax.set_xlabel("Annual salary", fontsize=20)
plt.show()
```
